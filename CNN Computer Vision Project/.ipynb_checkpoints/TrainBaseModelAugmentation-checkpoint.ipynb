```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import custom_module as cc #My custom module
```

## Data Gens and Loading Data PLUS Augmentation

```python
train_datagen = ImageDataGenerator(rescale=1./255, 
                                   horizontal_flip=True, #Added three augments. Mirroring, Altering brightness, and Channel Shift.
                                   brightness_range=[0.8, 1.2], #Lung CT images are generally near-semetrical, so I chose augments that won't affect that.
                                   channel_shift_range=0.3)     # The images in my dataset already had some varying brightness, so I included that in my augments.
                                                            # Used channel shift to hopefully allow some features to stand out that were previously ignored.


valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_dataset = cc.data_loader(train_datagen,'archive (13)/Data/train/') #Using my data loader module
valid_dataset = cc.data_loader(valid_datagen,'archive (13)/Data/valid/')
test_dataset = cc.data_loader(test_datagen,'archive (13)/Data/test/')
```

```python
cc.show_images(train_dataset) #Visualizing the augmentations
```

```python
# Shifts in brightness, channel shifts, and mirroring present.
```

## Training Model on Augmented Data

```python
#Here we go, again.
import keras

base_model = keras.applications.EfficientNetV2B0(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
    include_preprocessing=True,
)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

output = Dense(4, activation='softmax')(base_model.output)

model = Model(inputs=base_model.input, outputs=output)

model.compile(loss='categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])

history = model.fit(train_dataset,
                    epochs=5,  
                    batch_size=32, 
                    validation_data=valid_dataset)
```

```python
plt.plot(range(len(history.history["loss"])),history.history["loss"],label="Training Loss")
plt.plot(range(len(history.history["val_loss"])),history.history["val_loss"],label="Validation Loss")
plt.legend()
```

```python
plt.plot(range(len(history.history["accuracy"])),history.history["accuracy"],label="Training Accuracy")
plt.plot(range(len(history.history["val_accuracy"])),history.history["val_accuracy"],label="Validation Accuracy")
plt.legend()
```
