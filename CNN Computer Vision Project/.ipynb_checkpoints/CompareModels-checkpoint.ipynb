## Loading data

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_dataset = cc.data_loader(test_datagen,'archive (13)/Data/test/')
```

## Roc curves

```python
from sklearn.metrics import roc_curve, auc
```

```python
pred_base = base_model.predict(test_dataset)
pred_eff = eff_model.predict(test_dataset)
```

```python
fpr_base = dict()
tpr_base = dict()
roc_auc_base = dict()

num_classes=4

for i in range(num_classes):
    fpr_base[i], tpr_base[i], _ = roc_curve((test_dataset.labels == i).astype(int), pred_base[:, i])
    roc_auc_base[i] = auc(fpr_base[i], tpr_base[i])


for i in range(num_classes):
    plt.plot(fpr_base[i], tpr_base[i], label=f'Base Model (Class {i}) (AUC = {roc_auc_base[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class Base Model')
plt.legend()
plt.show()
```

```python
fpr_eff = dict()
tpr_eff = dict()
roc_auc_eff = dict()

for i in range(num_classes):
    fpr_eff[i], tpr_eff[i], _ = roc_curve((test_dataset.labels == i).astype(int), pred_eff[:, i])
    roc_auc_eff[i] = auc(fpr_eff[i], tpr_eff[i])


for i in range(num_classes):
    plt.plot(fpr_eff[i], tpr_eff[i], label=f'eff Model (Class {i}) (AUC = {roc_auc_eff[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class EfficientNet Model')
plt.legend()
plt.show()
```
